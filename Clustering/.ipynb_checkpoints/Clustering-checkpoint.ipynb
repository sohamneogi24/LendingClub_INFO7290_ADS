```python
# work on combined dataset

import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.externals import joblib
from sklearn.cluster import KMeans
from sklearn import preprocessing

%matplotlib inline

```

```python
#reading the file

accepted_df = pd.read_csv('../Data/combined.csv')
```

```python
accepted_df.shape
```

```python
#threshold of 70%
accepted_df = accepted_df.dropna(axis=1, thresh=0.30 * accepted_df.shape[0])
accepted_df = accepted_df.dropna(axis=0, thresh = 0.30 * (accepted_df.shape[1]))
accepted_df.shape
```

```python
#rejecting fico score low below 660
accepted_df = accepted_df[accepted_df['fico_range_low'] > 660]

accepted_df.shape
```

```python
#checking the unique grades

accepted_df.grade.unique()
```

```python
#replacing the % in int_rate

accepted_df['int_rate'] = accepted_df['int_rate'].map(lambda x : x.replace('%', ''))
accepted_df['int_rate'] = accepted_df['int_rate'].map(float)
```

```python
grade_interest = accepted_df[['grade', 'int_rate']]
grade_interest_mean = grade_interest.groupby(by='grade').mean().reset_index()

```

```python
#plotting grade v int_rate

grade_interest_mean.plot(kind='bar', x='grade', y='int_rate')

```

```python
#subgrade -> fico score, loan amnt, number of open accounts, loan term, purpose, state, debt_settlement_flag, dti

```

```python
accepted_df['risk_score'] = (accepted_df['fico_range_low'] + accepted_df['fico_range_high']) / 2
```

```python
accepted_df['term'].unique()
```

```python
# dti - elimination dti - debt / income , sub_grade
accepted_df_subset = accepted_df[['risk_score', 'loan_amnt', 'term', 'purpose', 'addr_state', 'debt_settlement_flag', 'dti', 'annual_inc', 'sub_grade', 'int_rate']]
```

```python
# dropping the nulls
accepted_df = accepted_df[accepted_df['risk_score'].notnull() & accepted_df['loan_amnt'].notnull() & accepted_df['term'].notnull() & accepted_df['purpose'].notnull() & accepted_df['addr_state'].notnull() & accepted_df['debt_settlement_flag'].notnull() & accepted_df['dti'].notnull() & accepted_df['annual_inc'].notnull() & accepted_df['grade'].notnull() & accepted_df['sub_grade'].notnull() & accepted_df['int_rate'].notnull()]
accepted_df['term'] = accepted_df['term'].map({' 36 months': 36, ' 60 months' : 60})
accepted_df.shape
```

```python
#Lending club isnt giving the load for people of states IA and WV

accepted_df = accepted_df[accepted_df['risk_score'] > 660]
accepted_df = accepted_df[accepted_df['dti'] <= 35]
accepted_df_clustering = accepted_df[accepted_df['addr_state'] != 'IA']
accepted_df_clustering = accepted_df[accepted_df['addr_state'] != 'WV']
accepted_df_clustering_final = accepted_df_clustering[['risk_score', 'loan_amnt', 'term', 'purpose', 'addr_state', 'debt_settlement_flag', 'dti', 'annual_inc', 'grade', 'sub_grade', 'int_rate']]
accepted_df_clustering_final.shape
```

```python
sns.boxplot(data=accepted_df_clustering_final['int_rate'])
```

```python
accepted_df = accepted_df[accepted_df['int_rate'] <=25]
accepted_df_clustering_final = accepted_df_clustering_final[accepted_df_clustering_final['int_rate'] <=25]
```

```python
sns.boxplot(data=accepted_df_clustering_final['loan_amnt'])
```

```python
accepted_df = accepted_df[accepted_df['loan_amnt'] <= 38000]
accepted_df_clustering_final = accepted_df_clustering_final[accepted_df_clustering_final['loan_amnt'] <=38000]
```

```python
#checking the columns in the data frame
accepted_df_clustering_final.columns
```

```python
#selecting the features
accepted_df_clustering_final_training = accepted_df_clustering_final[['risk_score', 'loan_amnt', 'term', 'purpose', 'addr_state', 'debt_settlement_flag', 'dti', 'annual_inc', 'sub_grade', 'int_rate']]
accepted_df_clustering_final_training.shape
```

```python
accepted_df_clustering_final.columns
```

```python
#creating dummy variables for purpose and sub_grade
purpose_dummies = pd.get_dummies(accepted_df_clustering_final_training['purpose'])
sub_grade_dummies = pd.get_dummies(accepted_df_clustering_final_training['sub_grade'])


accepted_clustering_with_dummies = accepted_df_clustering_final_training.join(purpose_dummies)
accepted_clustering_with_dummies = accepted_clustering_with_dummies.join(sub_grade_dummies)

accepted_clustering_with_dummies.drop('purpose', 1, inplace=True)
accepted_clustering_with_dummies.drop('addr_state', 1, inplace=True)
accepted_clustering_with_dummies.drop('debt_settlement_flag', 1, inplace=True)
accepted_clustering_with_dummies.drop('sub_grade', 1, inplace=True)

#writing out a csv
accepted_clustering_with_dummies.to_csv('../Data/clustering_dataset.csv', index = False)
```

```python
accepted_clustering_with_dummies_temp = accepted_clustering_with_dummies.drop(['int_rate'], axis = 1)
```

```python
#normalizing data for k-means
normalized_data = preprocessing.normalize(accepted_clustering_with_dummies_temp)
```

```python
normalized_data.shape
```

```python
# K means clustering
clusters = [4,6,8,10,12]
def cluster_data(k, dataset) :
    kmeans = KMeans(n_clusters=k, random_state=0, verbose=1)
    kmeans.fit(dataset)
    return kmeans.inertia_
    

inertia_values  = [cluster_data(el, normalized_data) for el in clusters]
inertia_values
```

```python
#plotting elbow curve to check the optimal cluster

elbow_curve = pd.DataFrame()
elbow_curve['k'] = clusters
elbow_curve['inertia'] = inertia_values

```

```python
elbow_curve.head()
```

```python
elbow_curve.plot(kind='line', x='k', y='inertia')
```

```python
# selecting 8 clusters as per the elbow graph

kmeans = KMeans(n_clusters=8, random_state=0, verbose=1)
kmeans.fit(normalized_data)
```

```python
#dumping the model
joblib.dump(kmeans, 'k-means.pkl')
```

```python
#concat the data frame and create a label

df1 = pd.DataFrame(accepted_clustering_with_dummies)
df2 = pd.DataFrame(kmeans.labels_, columns = ['kmean.label'])
df_Kmeans_cluster = pd.concat([df1, df2], axis = 1)
```

```python
#removing null values

df_Kmeans_cluster = df_Kmeans_cluster[df_Kmeans_cluster['risk_score'].notnull() &
                                      df_Kmeans_cluster['loan_amnt'].notnull() &
                                      df_Kmeans_cluster['term'].notnull() &
                                      df_Kmeans_cluster['dti'].notnull() &
                                      df_Kmeans_cluster['annual_inc'].notnull() &
                                      df_Kmeans_cluster['int_rate'].notnull()]
```

```python
#Grouping into clusters based on K means label

group_by_kmeans = df_Kmeans_cluster.groupby(['kmean.label'])
group_by_kmeans_df = list(group_by_kmeans)
```

```python
#writing out csv for k means clusters 
count = 0
kMeans_cluster = []
for kMeans_group in group_by_kmeans_df:
    df_Kmeans_cluster = group_by_kmeans_df[count][1]
    kMeans_cluster.append(df_Kmeans_cluster)
    print("K means Cluster", count, "of ", len(df_Kmeans_cluster), " rows")
    csv_name = "../Data/KMeansCluster/K-means_Cluster_" + str(count) + ".csv"
    df_Kmeans_cluster.to_csv(csv_name, index = False)
    count+=1
```
